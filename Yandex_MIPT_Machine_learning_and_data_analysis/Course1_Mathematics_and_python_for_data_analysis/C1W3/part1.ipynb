# Задача 1. Минимизация гладкой функции

```python
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import math
import numpy as np
```

## Часть 1. 
Рассмотрим все ту же функцию из задания по линейной алгебре: f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2), но теперь уже на промежутке [1, 30]

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

data_x = np.arange(1, 31, 0.1)
data_y = [f(x) for x in data_x]

plt.plot(data_x, data_y)
plt.show()
```

## Часть 2.
В первом задании будем искать минимум этой функции на заданном промежутке с помощью scipy.optimize.

```python
# попробуйте найти минимум, используя стандартные параметры в функции scipy.optimize.minimize 
#(т.е. задав только функцию и начальное приближение).
x = 10
res = minimize(f, x)
print(res.fun)
```

```python
# укажите в scipy.optimize.minimize в качестве метода BFGS, запустите из начального приближения x=2
x = 30
res = minimize(f, x, method='BFGS')
print(res.fun)
# значения миннимума при x = 2, x = 30 и будут ответом на задание
```

```python
# посмотрим, меняется ли результат, при изменении начального приближения
x = range(30)
res = [minimize(f, s, method='BFGS').fun for s in x]
print(res)
```

### Вывод
При подборе точек начального приближения от 0 до ~11, можно увидеть, что алгоритмы "проваливаются" в точку локального миннимума (1.74). Отсюда можно сделать вывод, что данные алгоритмы не подходят для поиска глобального миннимума.
