# Задача 2. Глобальная оптимизация

```python
from scipy.optimize import differential_evolution
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import math
import numpy as np
```

## Часть 1.
Теперь попробуем применить к той же функции f(x) метод глобальной оптимизации — дифференциальную эволюцию.

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

data_x = np.arange(1, 31, 0.1)
data_y = [f(x) for x in data_x]

plt.plot(data_x, data_y)
plt.show()
```

```python
# запустите поиск минимума функции f(x) с помощью дифференциальной эволюции на промежутке [1, 30]
res = differential_evolution(f, bounds=[(1, 30)])
print(res.fun)
# полученное значение функции в точке минимума - ответ в задаче 2
```

# Задача 3. Минимизация негладкой функции

Теперь рассмотрим функцию h(x) = int(f(x)) на том же отрезке [1, 30], т.е. теперь каждое значение f(x) приводится к типу int и функция принимает только целые значения.

```python
def h(x):
    return int(f(x))

data_y_h = [h(x) for x in data_x]   
plt.plot(data_x, data_y_h)
plt.show()
```

## Часть 1.
Попробуйте найти минимум функции h(x) с помощью BFGS, взяв в качестве начального приближения x=30. Получившееся значение функции – ваш первый ответ в этой задаче.

```python
x = 30
res = minimize(h, x, method='BFGS')
print(res.fun)
```

## Часть 2.
Теперь попробуйте найти минимум h(x) на отрезке [1, 30] с помощью дифференциальной эволюции. Значение функции h(x) в точке минимума – это ваш второй ответ в этом задании.

```python
res = differential_evolution(h, bounds=[(1, 30)])
print(res.fun)
```
